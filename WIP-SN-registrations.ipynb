```python
subtensor = bt.subtensor('archive')
```

```python
blocks = range(2800000, 2828762, 100)
subnet_emissions = {}
for block in blocks:
    emission = float(subtensor.get_emission_value_by_subnet(10, block))
    print(emission)
    subnet_emissions[block] = emission

emissions_df = pd.DataFrame(subnet_emissions, index=[0]).T
```
